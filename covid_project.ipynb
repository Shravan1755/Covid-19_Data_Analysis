```python
# Import the libraries..
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the datasets..
dataset = pd.read_excel('C:\\Users\\choud\\OneDrive\\Downloads\\Tutedude resources\\Tutedude Resourses\\covid19_Confirmed_dataset.xlsx')
dataset.head()
```

```python
dataset.shape
```

```python
# Delete the useless columns..
df = dataset.drop(["Lat", "Long"], axis=1, inplace=True)
dataset.head()
```

```python
# Aggregate the rows by the country..
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.shape
```

```python
# Check data types and sample data
print("Data Types:")
print(corona_dataset_aggregated.dtypes)
print("\nSample data for China:")
print(corona_dataset_aggregated.loc["China"].head())
```

```python
# Convert data to numeric and handle missing values
for country in ["China", "Spain", "Italy"]:
    # Convert to numeric, coerce invalid values to NaN
    corona_dataset_aggregated.loc[country] = pd.to_numeric(
        corona_dataset_aggregated.loc[country], 
        errors='coerce'
    )
    
# Create the visualization with cleaned data
# corona_dataset_aggregated.loc["China"].plot()
# corona_dataset_aggregated.loc["Spain"].plot()
# corona_dataset_aggregated.loc["Italy"].plot()
# plt.legend(["China", "Spain", "Italy"])
# plt.title("COVID-19 Cases by Country")
# plt.xlabel("Date")
# plt.ylabel("Number of Cases")
# plt.show()
```

```python
# Convert all columns to numeric first
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    # Convert to numeric and handle any invalid values
    numeric_data = pd.to_numeric(corona_dataset_aggregated.loc[c], errors='coerce')
    
    # Calculate max difference
    max_infection_rates.append(numeric_data.diff().mean())

corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
print(max_infection_rates)
```

```python
# Create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])
corona_data
```
